```python
def sample_data(data = None , n= len(data)) : 
    output = data[list(np.random.choice(len(data) , n))]
    return output

input_dim = np.shape(data)[1]
zdim = 40
batch_size = 4000

print("input_dim : {}  , z_dim : {} Batch size : {}".format(input_dim , zdim , batch_size))
```

```python
scope = ["GAN/shared" , 
        "GAN/fac" , 
        "GAN/num" , 
        "DIS/fac" , 
        "DIS/num" , 
        "DIS/total"]
```

```python
hsize_shared = [15 , 15]

def G_shared( Z , hsize = hsize_shared , reuse = tf.AUTO_REUSE) : 
    with tf.variable_scope(scope[0] , reuse = reuse) : 
        with slim.arg_scope( [slim.fully_connected ] , 
                            activation_fn = tf.nn.leaky_relu ,
                            weights_initializer = tf.variance_scaling_initializer()) : 
                shared = slim.stack(Z , slim.fully_connected , hsize )
        
        return shared
```

```python
def G_fac( Z , hsize = hsize_shared , reuse = tf.AUTO_REUSE) : 
    with tf.variable_scope(scope[1] , reuse = reuse) : 
        with slim.arg_scope([slim.fully_connected] , activation_fn = tf.nn.leaky_relu , 
                           weights_initializer = tf.variance_scaling_initializer()) : 
            shared = slim.stack( Z  , slim.fully_connected , hsize )
        output = slim.fully_connected(shared , len(fac_var))
        
        return output
```

```python
def G_num( Z , hsize = hsize_shared , reuse = tf.AUTO_REUSE) : 
    with tf.variable_scope(scope[2] , reuse = reuse) : 
        with slim.arg_scope([slim.fully_connected] , activation_fn = tf.nn.leaky_relu , 
                           weights_initializer = tf.variance_scaling_initializer()) : 
            shared = slim.stack( Z  , slim.fully_connected , hsize )
            
        output = slim.fully_connected(shared , len(num_var))
        
        return output
```

```python
hsize_dis = [55,38,25]
def D_fac(X, hsize=hsize_dis  ,reuse=tf.AUTO_REUSE ):
    with tf.variable_scope( scope[3] ,reuse=reuse):
        h = tf.layers.dense(X,hsize[0], activation=tf.nn.leaky_relu)
        for i in np.arange( len(hsize)-1 )  : 
            h = tf.layers.dense(h,hsize[i+1],activation=tf.nn.leaky_relu ,name="fac_w_"+str(i) )
        out = tf.layers.dense(h,1) # activation=None
        return out
```

```python
def D_num(X, hsize=hsize_dis  ,reuse=tf.AUTO_REUSE ):
    with tf.variable_scope( scope[4] ,reuse=reuse):
        h = tf.layers.dense(X,hsize[0], activation=tf.nn.leaky_relu)
        for i in np.arange( len(hsize)-1 )  : 
            h = tf.layers.dense(h,hsize[i+1],activation=tf.nn.leaky_relu ,name="num_w_"+str(i) )
        out = tf.layers.dense(h,1) # activation=None
        return out
    
```

```python
def D_total(X, hsize=hsize_dis  ,reuse=tf.AUTO_REUSE ):
    with tf.variable_scope( scope[5] ,reuse=reuse):
        h = tf.layers.dense(X,hsize[0], activation=tf.nn.leaky_relu)
        for i in np.arange( len(hsize)-1 )  : 
            h = tf.layers.dense(h,hsize[i+1],activation=tf.nn.leaky_relu ,name="total_w_"+str(i) )
        out = tf.layers.dense(h,1) # activation=None
        return out
    
```

```python
X_total = tf.placeholder(tf.float32,[None,input_dim ])
X_fac = tf.placeholder(tf.float32,[None, len(fac_var) ])
X_num = tf.placeholder(tf.float32,[None, len(num_var) ])
Z = tf.placeholder(tf.float32,[None, zdim ])
hint_sk = tf.placeholder(tf.float32,[1, len(num_var) ])
hint_ku = tf.placeholder(tf.float32,[1, len(num_var) ])
```

```python
def tf_skewness(X) : 
    numer = tf.reduce_mean(tf.pow(tf.subtract(X ,tf.reduce_mean(X , axis=0)),3), axis=0)
    denom = tf.pow(tf.reduce_mean(tf.square(tf.subtract(X ,tf.reduce_mean(X , axis=0))), axis=0),1.5)
    output = tf.divide(numer , denom)
    return output

def tf_kurt(X) : 
    numer = tf.reduce_mean(tf.pow(tf.subtract(X ,tf.reduce_mean(X , axis=0)),4) , axis = 0 )
    denom = tf.pow(tf.reduce_mean(tf.pow(tf.subtract(X ,tf.reduce_mean(X , axis=0) ),2),axis=0),2)
    output = tf.subtract(tf.divide(numer , denom),3.0)
    return output


```

```python
shared_layer = [22,19]
SHARE = G_shared(Z , hsize = shared_layer )
```

```python
G_FAC__ = G_fac(SHARE)
G_NUM__ = G_num(SHARE)
G_concat = tf.concat([G_FAC__ , G_NUM__] ,axis = 1 )
```

```python
residual_part = tf.reduce_sum(tf.square(tf.subtract(X_fac , G_FAC__))) / int(batch_size/4) 
```

```python
Real_fac = D_fac(X_fac)
Real_num = D_num(X_num)
Real_total = D_total(X_total)

Gene_fac = D_fac(G_FAC__)
Gene_num = D_num(G_NUM__ )
Gene_total = D_total(G_concat)
```

```python
#D_sk = tf_skewness(X_num)
G_sk = tf_skewness(G_NUM__)
stat_sk = tf.reduce_mean(tf.square(tf.subtract(hint_sk , G_sk)))

#D_ku = tf_kurt(X_num)
G_ku = tf_kurt(G_NUM__)
stat_ku = tf.reduce_mean(tf.square(tf.subtract(hint_ku , G_ku)))
```

# WGAN_GP


```python

def gradient_penalty(real, fake, f):
    alpha = tf.random_uniform(
        shape=[batch_size ,1], 
        minval=0.,
        maxval=1.
    )
    differences = fake- real
    interpolates = real + (alpha*differences)
    gradients = tf.gradients(f(interpolates), [interpolates])[0]
    slopes = tf.sqrt(tf.reduce_sum(tf.square(gradients), reduction_indices=[1]))
    gp = tf.reduce_mean((slopes-1.)**2)
    return gp

##fac
wd_fac = tf.reduce_mean(Real_fac) - tf.reduce_mean(Gene_fac)
gp_fac = gradient_penalty(X_fac, G_FAC__, D_fac)
d_loss_fac = -wd_fac + gp_fac * 10.0
g_loss_fac = -tf.reduce_mean(Gene_fac)

##num
wd_num = tf.reduce_mean(Real_num) - tf.reduce_mean(Gene_num)
gp_num = gradient_penalty(X_num , G_NUM__, D_num)
d_loss_num = -wd_num + gp_num * 10.0
g_loss_num = -tf.reduce_mean(Gene_num)

##total
wd_total = tf.reduce_mean(Real_total) - tf.reduce_mean(Gene_total)
gp_total = gradient_penalty(X_total, G_concat , D_total)
d_loss_total = -wd_total + gp_total * 10.0
g_loss_total = -tf.reduce_mean(Gene_total)



Dloss = d_loss_fac + d_loss_num + d_loss_total + residual_part 
Gloss = g_loss_fac + g_loss_num + g_loss_total + stat_sk + stat_ku
```

```python

gen_vars = []
for i in np.arange(3) : 
    output = tf.get_collection(tf.GraphKeys.GLOBAL_VARIABLES,scope=scope[i])
    for j in output :
        gen_vars.append(j)
        
dis_vars = []
for i in np.arange(4,len(scope)) : 
    output = tf.get_collection(tf.GraphKeys.GLOBAL_VARIABLES,scope=scope[i])
    for j in output :
        dis_vars.append(j)

```

```python

global_step = tf.Variable(0, trainable=False)
optimizer = tf.train.AdamOptimizer(learning_rate=
  clr.cyclic_learning_rate(global_step=global_step, mode='triangular2' , gamma=.997 ,
                           learning_rate=0.0009,max_lr=0.07,step_size=30.))


# G Train step
gen_step = optimizer.minimize(Gloss,var_list = gen_vars, global_step=global_step) 
# D Train step
disc_step = optimizer.minimize(Dloss,var_list = dis_vars, global_step=global_step) 
```

```python
import warnings
warnings.filterwarnings('ignore')

sess = tf.Session()
tf.global_variables_initializer().run(session=sess)

nd_steps = 5
ng_steps = 10

iteration , dloss , gloss = 0 , 0 ,0
output = pd.DataFrame({"iter" : [iteration] , "dloss" : [dloss] , "gloss" : [gloss]})
global_step_count = 0
rates = []
for i in range(25001):
    X_batch = sample_data(data = data , n=batch_size)
    Z_batch = sample_Z(batch_size , zdim )
    
    global_step_count += 1
    
    assign_op = global_step.assign(global_step_count)

    for _ in range(nd_steps):
        global_step_count += 1
        assign_op = global_step.assign(global_step_count)
        rates_ , _, dloss , res  = sess.run([assign_op , disc_step, Dloss , residual_part  ], feed_dict={X_total : X_batch, 
                                             X_fac : X_batch[: , list(np.arange(0 , len(fac_var))) ] , 
                                             X_num : X_batch[: , list(np.arange(len(fac_var), len(col) )) ] , 
                                             Z: Z_batch})
        rates.extend([rates_])

    for _ in range(ng_steps):
        global_step_count += 1
        assign_op = global_step.assign(global_step_count)
        rates_ , _, gloss ,stat_sk2 ,stat_ku2 = sess.run([assign_op, gen_step, Gloss ,stat_sk ,stat_ku ], feed_dict={Z: Z_batch , hint_sk : hint_skew , hint_ku : hint_kurt })
        rates.extend([rates_])
        
    
    dcount = 0 
    while dloss > 9 : 
        dcount += 1
        global_step_count += 1
        assign_op = global_step.assign(global_step_count)
        rates_ , _, dloss , res = sess.run([assign_op , disc_step, Dloss , residual_part ], feed_dict={X_total : X_batch, 
                                             X_fac : X_batch[: , list(np.arange(0 , len(fac_var))) ] , 
                                             X_num : X_batch[: , list(np.arange(len(fac_var), len(col) )) ] , 
                                             Z: Z_batch})
        rates.extend([rates_])
        
        if dcount > 510 : 
            if i % 10 == 0  : 
                print(" {} 번째 탈출".format(i))
                clear_output(wait= True)
            break
    gcount = 0 
    while gloss > 9 :
        gcount += 1 
        for _ in range(ng_steps):
            global_step_count += 1
            assign_op = global_step.assign(global_step_count)
            rates_ , _, gloss ,stat_sk2 ,stat_ku2 = sess.run([assign_op ,gen_step, Gloss,stat_sk ,stat_ku ], feed_dict={Z: Z_batch , hint_sk : hint_skew , hint_ku : hint_kurt })
            rates.extend([rates_])
        if gcount > 510 :
            break
    
        
    
    iteration +=1
    output1 = pd.DataFrame({"iter" : [iteration] , "dloss" : [dloss] , "gloss" : [gloss]})
    output = output.append(output1)
    output = output[output.iter>0]
    if i% 50 == 0:
        dloss = round( np.float64(dloss),4)
        gloss = round( np.float64(gloss),4)
        res = round( np.float64(res),4)
        show_plot(output = output , row = 7 , ncol=7)
        
        plt.xlabel('iterations (epochs)')
        plt.ylabel('learning rate')
        plt.plot(range(global_step_count), rates)
        plt.savefig("./rates_plot.png")
        plt.show()
```
